```python
import pandas as pd
```

```python
filename = '/content/athleteEventsNoPersonal.csv'
df = pd.read_csv(filename)
```

```python
df.head(20)
```

```python
df.shape # Filas y columnas del Dataframe
```

```python
df.isna().sum() # Le faltan datos a columnas 'Age', 'Height' y 'Medal'. En cuanto a la falta de datos importante en 'Medal' se debe principalmente que no todos los atletas consiguen medalla en cada una de sus participaciones
```

```python
dato_London = df['City'] == 'London'
df.loc[dato_London, :] # London tiene 3.370 filas
```

```python
df['Age'].min() # El Atleta más joven tenía 11 años al momento de competir
```

```python
df['Sport'].value_counts()
```

```python
dato_tug_of_war = df['Sport'] == 'Tug-Of-War'
df.loc[dato_tug_of_war, :] # Cantidad totoal de atletas en el deporte'Tug-Of-War'
```

```python
filtro = (df['Sport']=='Tug-Of-War') & (df['Team']=='Sweden')
filtro.sum() # Aplicado el filtro, se confirma que los 6 atletas que han participado en este deporte, son de Suecia.
```

```python
medallas = df['Medal']=='Gold'
df.loc[medallas, :]
```

```python
filtro_2 = (df['Medal']=='Gold') & (df['Age'] > 40)
filtro_2.value_counts() # En 66 ocasiones mayores de 40 años consiguieron las medallas de oro
```

```python
altura = df['Height'].mean() # Altura promedio de todas las filas 
altura
```

```python
df.groupby('Event')['Height'].mean() # Promedio de altura, según evento.
```

```python
df.drop_duplicates(subset = 'ID', keep = 'first')
```

```python
df.groupby('ID')['Height'].mean() # Se saca la altura promedio de los atletas, sin que este se repita. 
```
